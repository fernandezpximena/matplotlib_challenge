```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
```

```python
clinical = pd.read_csv("clinicaltrial_data.csv", encoding="UTF-8")
mouse= pd.read_csv("mouse_drug_data.csv", encoding="UTF-8")
```

```python
clinical.head()
```

```python
mouse.head()
```

```python
clinical.plot(kind="scatter",x="Timepoint", y="Tumor Volume (mm3)",
              title="Volume Changes")
plt.show()
```

```python
clinical.plot(kind="scatter",x="Tumor Volume (mm3)", y="Metastatic Sites",
              title="Volume Changes")
plt.show()
```

```python
merge_table = pd.merge(clinical, mouse, on="Mouse ID")
merge_table
```

```python
drug_stats = merge_table.groupby(["Drug"])
drug_stats.count().head()
```

```python
drug_stats.plot(kind="bar",x="Drug", y="Tumor Volume (mm3)",
              title="Drug Comparative")
plt.show()
```
